In this assignment students have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
from sklearn.cluster import KMeans
```

```python
#loading the racoon grey scale image
image = scipy.misc.face(gray=True)
```

```python
image.shape
```

```python
#visualize the grey scale image
plt.imshow(image, cmap=plt.cm.gray)
plt.show()
```

```python
#Assigning the rows and columns
rows = image.shape[0]
cols = image.shape[1]
```

```python
#reshaping the image from 2D to 1D 
reshaped_image = image.reshape(rows*cols,1)
```

```python
reshaped_image.shape
```

```python
# Applying KMeans
kmeans = KMeans(n_clusters = 5, n_init = 1, max_iter = 200, verbose = 1)
kmeans.fit(reshaped_image)
```

```python
# 5 cluster centroids
print(kmeans.cluster_centers_) 
```

```python
# converting in to array and dtype
clusters = np.asarray(kmeans.cluster_centers_, dtype = np.uint8)
labels = np.asarray(kmeans.labels_, dtype = np.uint8)
```

```python
# reshaping back in to image
labels = labels.reshape(rows,cols)
```

```python
#saving the compressed image
plt.imsave("Compressed_racoon_image.png", labels)
```

```python
#visulaizing the compressed image
compressed_image = plt.imread('Compressed_racoon_image.png')
plt.imshow(compressed_image)
plt.show()
```
