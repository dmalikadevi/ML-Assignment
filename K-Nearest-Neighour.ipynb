```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
```

```python
pd.set_option('display.max_columns',None)
nba = pd.read_csv("E:/Malika/Assignment/nba_2013.csv")
```

```python
nba.isnull().sum()
```

```python
# Fill null value columns with its mean value
nba["ft."] = nba["ft."].fillna(nba["ft."].mean())
nba["efg."] = nba["efg."].fillna(nba["efg."].mean())
nba["x2p."] = nba["x2p."].fillna(nba["x2p."].mean())
nba["x3p."] = nba["x3p."].fillna(nba["x3p."].mean())
nba["fg."] = nba["fg."].fillna(nba["fg."].mean())
```

```python
# New data frame with few dropped columns 
nba_withdroppedcolumns = nba.drop(['player','age','pos','bref_team_id','season','season_end'], axis=1)
```

```python
nba_withdroppedcolumns.info()
```

```python
# Normalizing the x values
nba_new = nba_withdroppedcolumns.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
nba_new.head()
```

```python
x = nba_new
y = nba_new['pts']
```

```python
# spliting data randomly for training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# To find the increasing neighbour K values and their regression score

for k in range(10):
    k_value = k + 1
    knn = KNeighborsRegressor(n_neighbors = k_value)
    knn.fit(x_train, y_train) 
    y_pred = knn.predict(x_test)
    print ("Regression score is:",format(metrics.r2_score(y_test, y_pred),'.4f'), "for k_value:", k_value)
```
